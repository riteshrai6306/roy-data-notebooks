# Data Cleaning

# 1. Dealing with Null Values

```python
#to check the null values in entire dataset
df.isnull().sum()
```

There are two ways to deal with Null values:-
1. Either Drop the rows or null values as per your requirement
2. Or Fill those null values with the default ones

```python
#to remove the null values from the dataframe not from the original one.
df.dropna()
```

```python
#you can pass an argument to drop null values
df.dropna(how = 'any') #any row that has any null values (So, you can drop the null values as per your requirements)
```

```python
df.dropna(how = 'all') #if all the values in any row are null then we drop
```

# How to deal with missing values in a dataset
>> getting rid of those values using "dropna(how = 'any'/ how = 'all')"
>> filling those missing values "fillna()"

You can fill missing values using mean() or median() or any default value as per your requirement.

```python
#you can fill the mising values using mean() or any default values
df['Age'].fillna(df['Age'].mean())
```

```python
#filling the salary null value using median value
df['Salary'].fillna(df['Salary'].median())
```

```python
df
```

```python
#fillna() has two methods - Backwordfill & forwardfill
df['Age'].fillna(method = 'ffill')
```

```python
df['Age'].fillna(method = 'bfill')
```

***`Imp:- However, using ffill or bfill to replace the null values can be tricky if previous or the next values are null`***

```python
#want to replace the particular values with another one
df['Name'].replace('Charlie', 'Rose')
```

```python
df['Name'] = df['Name'].replace('Charlie', 'Rose') #to reflect the changes in the original dataset
```

```python
df
```

# 2. Dealing with Duplicates data:-



```python
#to check the duplicats values
df_dup = df[df.duplicated()]
df_dup
```

```python
df.duplicated() #it return boolean values if True it's duplicated
```

```python
#to check the first & last duplicates
df_dup = df[df.duplicated(keep = 'first')]
df_dup
```

```python
#if you want to drop the duplicate values and want to keep the first one
df = df.drop_duplicates(keep = 'first')
```

```python
df
```

```python
#how we can deal with invalid values
#using 'lambda' and 'apply' function
df['Promoted Salary'] = df['Promoted Salary'].apply(lambda x: x/10 if x > 650000 else x)
```

```python
df
```

```python
#apply & lambda function - very useful as they let you perform and change single values of you columns
def multiplying_age(x):
  return x*2

df['Age'] = df['Age'].apply(multiplying_age)
```

```python
df
```

```python
#to get back to the original age
df['Age'] = df['Age'].apply(lambda x: x/2)
```

```python
df
```

That's how apply & lambda function work

# Joins & Merges

```python
#Join & merges
department_info = {
    "Dept":["HR", "IT", "Finance"],
    "Location": ["New York", "San Francisco", "Chicago"],
    "Manager":["Laura", "Steve", "Nina"]

}
```

```python
df2 = pd.DataFrame(department_info)
df2
```

```python
#Using concat to merget both the table
pd.concat([df, df2]) #it helps us to combine the table horizontel stack
```

```python
pd.concat([df, df2], axis = 1) #to merge colum wise
```

```python
pd.merge(df , df2, on = "Dept") #merger the column on the basis of common column so that it's not repeated two times
```
